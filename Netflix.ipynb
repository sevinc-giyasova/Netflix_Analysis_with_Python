```python
df["date_added"]=pd.to_datetime(df["date_added"], format="mixed", errors="coerce")
df
```

```python
df['duration_value']= df["duration"].str.extract(r'(\d+)').astype(float)
df['duration_type']= df["duration"].str.extract(r'([A-Za-z]+)').astype(str)
df.info()
```

```python
df.isnull().sum()
```

```python
df["title"]=df["title"].str.strip().str.title()
df["type"]=df["type"].str.strip()
df["country"]=df["country"].str.strip()
```

```python
df["country"]=df["country"].fillna("Unkown")
df["cast"]=df["cast"].fillna("Unkown")
df["director"]=df["director"].fillna("Unkown")
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df = df.dropna(subset=["date_added","duration"])
df.head()
```

```python
rating_groups = {
    'Kids': ['TV-Y', 'TV-Y7', 'TV-G', 'G', 'TV-Y7-FV'],
    'Teens': ['TV-PG', 'PG', 'PG-13', 'TV-14'],
    'Adults': ['TV-MA', 'R', 'NC-17', 'NR', 'UR']
}
def cat(x):
  for k ,v in rating_groups.items():
    if x['rating'] in v:
      return k
df=df.copy()
df['category']=df.apply(lambda x : cat(x),axis=1)
```

```python
#NÃ¶vÉ™ gÃ¶rÉ™ qruplaÅŸdÄ±rma
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type", palette="Set2")
plt.title("Movie vs TV show", fontsize=14)
plt.ylabel("Count")
plt.xlabel("Type")
plt.show()
```

```python
#YaÅŸ kateqoriyasÄ±na gÃ¶rÉ™ qruplaÅŸdÄ±rma

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="category", palette="Set2", hue="type")
plt.title("Content by Age Category", fontsize=14)
plt.xlabel("category")
plt.ylabel("Count")
plt.show()
```

```python
#AÃ§Ä±qlamada istifadÉ™ olunan sÃ¶zlÉ™rÉ™ gÃ¶rÉ™ qruplaÅŸdÄ±rma
from textblob import TextBlob
from wordcloud import WordCloud
df['sentiment'] = df['description'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['sentiment_label'] = df['sentiment'].apply(lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral'))

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment_label', palette=['green','red','gray'])
plt.title('Sentiment of Descriptions', fontsize=14)
plt.ylabel('Count')
plt.xlabel('Sentiment')
plt.show()
```

```python
# AÃ§Ä±qlama hissÉ™sindÉ™ É™n Ã§ox tÉ™krarlanan sÃ¶zlÉ™r
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(df["description"].dropna())
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# 2000 vÉ™ sonrasÄ± Ã¼Ã§Ã¼n Ã§É™kilÉ™n film saylarÄ±
df_after_2000=df[df["release_year"]>=2000]

plt.figure(figsize=(12,5))
sns.countplot(
    data=df_after_2000, x= 'release_year', palette="Set2",
    order= sorted(df_after_2000["release_year"].dropna().unique())
)

plt.xticks(rotation=90)
plt.title("Number of titles per year (2000 an After)", fontsize=14)
plt.ylabel("Count")
plt.xlabel("Years")
plt.show()
```

```python
#Top 10 filmlÉ™r Ã¶lkÉ™lÉ™r Ã¼zrÉ™ bÃ¶lgÃ¼sÃ¼
top_countries = df["country"].value_counts().head()

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Set2")
plt.title("Top 10 countries by Number of titles", fontsize=14)
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
#BÃ¼tÃ¼n illÉ™rÉ™ gÃ¶rÉ™ Ã§É™kilÉ™n film sayÄ±
plt.figure(figsize=(12,5))
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Netflix content count by year')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
#Janrlar Ã¼zrÉ™ top 15 filmlÉ™r
genres= df["listed_in"].str.split(",", expand=True).stack()
plt.figure(figsize=(12,6))
genres.value_counts().head(15).plot(kind="bar")
plt.title("Top 15 most popular movies")
plt.ylabel("Count")
```

```python
# FilmlÉ™rin mÃ¼ddÉ™tÉ™ gÃ¶rÉ™ paylanmasÄ±
movies = df[df['type']=='Movie']
plt.figure(figsize=(10,5))
sns.histplot(movies['duration_value'], bins=30, kde=True, palette="Set2")
plt.title('Time distribution of films')
plt.xlabel('Min')
plt.show()
```

```python
#Ã–lkÉ™lÉ™r Ã¼zrÉ™ Movie/TV Show bÃ¶lgÃ¼sÃ¼
top_countries = df['country'].value_counts().head(10).index
sub = df[df['country'].isin(top_countries)]

plt.figure(figsize=(12,6))
sns.countplot(data=sub, x='country', palette="Set2", hue='type')
plt.title('Movie / TV Show distribution by countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Reytinq vÉ™ mÉ™zmun nÃ¶vÃ¼ Ã¼zrÉ™ paylanma
plt.figure(figsize=(10,6))
cross = df.pivot_table(index='rating', columns='type', aggfunc='size', fill_value=0,observed=False)
sns.heatmap(cross, annot=True, fmt='d')
plt.title('Rating vÉ™ Content Type Heatmap')
plt.show()
```
